```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, Y = make_blobs(n_samples=200, centers=2, 
                  random_state=0, cluster_std=0.40)
```

```python


# plotting scatters  
#  x and y coordinates are stored in one single 2D array called X: X[:,0] are simply the x coordinates, 
#  and X[:,1] the y coordinates.
#  c=Y indicating which flower the (x,y)-->(X[:,0],X[:,1]) coordinate point 

plt.scatter(X[:, 0], X[:, 1], c=Y, s=40, cmap='spring'); 
plt.show()  
```
